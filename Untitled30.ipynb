Ho:there is no significant difference between sales amt
H1:there is significance difference between sales

```python
fstat,pvalue=stats.f_oneway(data["Zone - A"],data["Zone - B"],data["Zone - C"])
print('the fstatistics is',fstat)
print('the pvalue is',pvalue)
if pvalue <.05:
    print("we reject null hypothesis")
else:
    print("we fail to reject null hypothesis")
```
